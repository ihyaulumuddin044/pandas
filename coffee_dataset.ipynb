```python
olympic['Age'] = olympic['Age'].fillna(0)  # Isi NaN dengan 0 untuk Age
olympic['Weight'] = olympic['Weight'].fillna(0)

def chategoryAthlete(row):
  if row['Age'] <= 20 and row['Weight'] <= 60:
    return 'Youth lightwight'
  elif row['Age'] <= 25 and row['Weight'] <= 70:
    return 'Youth middleweight'
  else:
    return 'Adult'

olympic['Chategory'] = olympic.apply(chategoryAthlete, axis=1)
```
