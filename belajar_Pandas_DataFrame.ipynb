# Melakukan akses data pada Dataframe

terdapat dua mekanisme untuk melakukan akses pada dataframe index-based selection dan lable-based selection.
kedua mekanisme akses ini mendahukan baris(row) baru di ikuti kolom (colomn)

> 1. index-based selection

metode ini dilakukan dengan memanfaatkan method iloc

```python
wine_df.iloc[0]
```

```python
wine_df.iloc[1,0]
```

slicing pada ilog

```python
wine_df.iloc[0,:7]
```

```python
wine_df.iloc[1:3, 0]
```

melewatkan list untuk melakukan akses data pada baris dan kolom tertentu

```python
wine_df.iloc[[0,2,4], 0]
```

>2.  lable-based selection

mekanisme dilakukan dengan method loc

```python
wine_df.loc[0, 'Alcohol']
```

```python
wine_df.loc[0:5, 'Alcohol']
```

```python
wine_df.loc[:, ['Alcohol', 'Phenols']].head()
```
