```python
import pandas as pd

data = {
    "Nama": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Usia": [25, 30, 35, 40, 45],
    "Kota": ["Jakarta", "Bandung", "Surabaya", "Jakarta", "Bandung"],
    "Pendapatan": [5000000, 6000000, 5500000, 7000000, 6500000],
    "Status": ["Single", "Married", "Married", "Single", "Single"]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
rata_rata_usia = df['Usia'].mean()

print('rata2 usia: ',rata_rata_usia)
```

```python
jakarta = df[df['Kota'] == 'Jakarta']
jakarta
```

```python
# total pendapatan semua yang single

penghasilan_single = df[df['Status'] == 'Single']['Pendapatan'].sum()
print('total pendapatan single: ',penghasilan_single)
```

```python
# menambahkan kolom baru

df['pajak'] = 10/100 * df['Pendapatan']
df
```

```python
# total pajak yang ada di bandung

pajak_bandung = df[df['Kota'] == 'Bandung']['pajak'].sum()
print('total pajak di bandung: ',pajak_bandung)
```

```python
pendaptan = df.loc[df["Pendapatan"] > 6000000, ['Nama',"Pendapatan"]]
print(f'pendapatan di atas 6 jt, \n{pendaptan}\n')
```

```python
# Tampilkan orang-orang yang memiliki status "Single" dan pendapatan lebih dari 6.000.000.

single = df.loc[(df["Status"] == "Single") & (df["Pendapatan"] > 6000000), ['Nama',"Status","Pendapatan"]]
single
```

```python
#  Tampilkan rata-rata pendapatan orang yang berstatus "Married" dan tinggal di "Bandung".

pendapatan = df.loc[(df["Status"] == "Married") & (df["Kota"] == "Bandung"), 'Pendapatan'].mean()
print('rata2 pendapatan yang married di bandung: ',pendapatan)
```
